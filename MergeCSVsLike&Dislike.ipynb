## Merge Like and Dislike blocks

Change the numeration of trials once they are separated in Like and Dislike conditions (avoid breaks in trial number)

```python
import numpy as np
import pandas as pd
from pandas.core.frame import DataFrame as DF
import seaborn as sns
sns.set(color_codes=True)
#np.random.seed(sum(map(ord, "distributions")))
from sklearn import linear_model  # packages for the logistic regression function to plot the logistic regression 
from sklearn.linear_model import LogisticRegression # packages for the logistic regression function to plot the logistic regression 
import scipy
from scipy import stats, integrate
from scipy.stats import mode
from scipy.stats.stats import pearsonr # Pearson's correlation
from copy import copy as copy
import operator as operator
import pylab
```

```python
#for binned data
# we use v1 that is the version that doesn't have the numeration corrected
data_L = pd.read_csv('/Users/bdmlab/Documents/GiTs/glamFF2018/data/FF2018_data/GlamDataFF2018_Like_Bin_v2.csv') 
data_D = pd.read_csv('/Users/bdmlab/Documents/GiTs/glamFF2018/data/FF2018_data/GlamDataFF2018_Dislike_Bin_v2.csv') 


data_L = data_L.apply(pd.to_numeric, errors='ignore')
data_D = data_D.apply(pd.to_numeric, errors='ignore')

data_L = data_L.drop(['Unnamed: 0'], 1)
data_D = data_D.drop(['Unnamed: 0'], 1)
```

## Invert choice  and change trial number for dislike block


```python
data_D
```

```python
data_D[['choice']] = data_D[['choice']].replace(0, -1)
data_D[['choice']] = data_D[['choice']].replace(1, 0)
data_D[['choice']] = data_D[['choice']].replace(-1, 1)

n_trials = data_D['trial'].unique()

for i in n_trials:
    data_D[['trial']] = data_D[['trial']].replace(i, 120+i)

data_D
```

# Merge Like and Dislike

```python
data_together = [data_L , data_D]
data_together = pd.concat(data_together)
```
