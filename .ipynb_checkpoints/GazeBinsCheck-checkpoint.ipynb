```python
#for binned data
# we use v1 that is the version that doesn't have the numeration corrected
data_1 = pd.read_csv('/Users/bdmlab/Documents/GiTs/FF2018/InputData/GlamDataFF2018_Like_Bin_v2_GAZE_BINS_TEST.csv')

data_1 = data_1.apply(pd.to_numeric, errors='ignore')

data_1 = data_1.drop(['Unnamed: 0'], 1)
```

```python
bins = np.linspace(-1, 1, 8+1)
bins
```

```python
data_1['DGaze'] = pd.cut(data_1['DGaze'],
                                               bins=bins, include_lowest=True,
                                               labels=bins[:-1])
data_1['left_chosen'] = data_1['choice'] == 0
```

```python
data_1
```

We don't have an equal number of members per bin in this case, therefore, when we calculate the mean value is not balanced.

```python
data_1['DGaze'].value_counts()
```

```python
data_1 = data_1.apply(pd.to_numeric, errors='ignore')
```

```python
partNums = pd.unique(data_1["subject"])
partNums
```

```python
binsNums = pd.unique(data_1["DGaze"])
binsNums
```

```python
partNums = pd.unique(data_1["subject"])
binsNums = pd.unique(data_1["DGaze"])

#new_fix_dataframe = data_fix.iloc[0:0]
# generate new dataset with the bins values
binsValues = []

for i in partNums:
    bin_row = []
    for j in binsNums:
         participantBin = data_1.loc[(data_1['subject'] == i) & (data_1['DGaze'] == j)]
         bin_value = participantBin['left_chosen'].mean() 
         bin_row.append(bin_value)        
    binsValues.append(bin_row)
df_BinsValues = pd.DataFrame(binsValues, columns = binsNums)
```

```python
df_BinsValues
```
